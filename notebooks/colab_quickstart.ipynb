```python
# Load secrets from Colab (optional)
from google.colab import userdata
import os

def _safe_get_secret(name: str):
    try:
        return userdata.get(name)
    except Exception:
        return None

openai_key = _safe_get_secret("OPENAI_API_KEY")
hf_token = _safe_get_secret("RHD_HF_TOKEN")

if openai_key:
    os.environ["OPENAI_API_KEY"] = openai_key
if hf_token:
    os.environ["RHD_HF_TOKEN"] = hf_token
    os.environ["HF_TOKEN"] = hf_token
    os.environ["HUGGINGFACE_HUB_TOKEN"] = hf_token
```

```python
!python -m rag_heuristics.cli ingest --source-dir data/raw
!python -m rag_heuristics.cli build-index
```

```python
# Run one seeded track (MDD)
# train now auto-exports paper-style discovered algorithms to reports/discovered_algorithms.md
!python -m rag_heuristics.cli train --problem-type single_machine_total_tardiness --iterations 60 --seed-strategy mdd

# Explicit export (single DB) with top 5 programs in paper-style format
!python -m rag_heuristics.cli export-discovered --problem-type single_machine_total_tardiness --top-k 5 --out reports/discovered_algorithms.md

# Compare EDD/SPT/MDD seeded tracks with RAG enabled
# compare-tracks now auto-exports paper-style report to reports/discovered_algorithms_by_seed_track.md
!python -m rag_heuristics.cli compare-tracks --problem-type single_machine_total_tardiness --iterations 60

# Optional: explicit per-track export to a custom file
!python -m rag_heuristics.cli export-discovered --problem-type single_machine_total_tardiness --tracks edd,spt,mdd --top-k 3 --out reports/discovered_algorithms_tracks.md

# Optional: full benchmark + ablations
!python -m rag_heuristics.cli evaluate --problem-type single_machine_total_tardiness --iterations 40

# Inspect generated reports
!ls -lah reports
```
